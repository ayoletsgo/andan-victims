**По итогам предварительного анализа датасета по заявленной теме можно сделать следующие выводы:**
*  В колонках содержатся следующие признаки:
1.   "Дата" - str формат даты фиксирования показателей
2.   "Максимальная температура" - float формат (после обработки ниже) максимальной температуры за сутки
3.   "Минимальная температура" - float формат (после обработки ниже) минимальной температуры за сутки
4.   "Средняя температура" - float формат (после обработки ниже) среднесуточной температуры
5.   "Атмосферное давление" - float формат (после обработки ниже) показателей атмосферного давления в единицах бар за сутки
6.   "Скорость ветра" - float формат (после обработки ниже) скорости ветра в м/с, зафиксированной в конкретный день
7.   "Осадки" - float формат (после обработки ниже) зафиксированных за сутки осадков в мм
8.   "Эффективная температура" - float формат (после обработки ниже) комплексного показателя (совокупное воздействие температуры и влажности), влияющего на состояние человека, в градусах.

*  Где это было возможно, произошла замена NaN-значений на предыдущие ввиду наличия прецедента использования в исследованиях датасетов погоды, например: https://towardsdatascience.com/weather-forecasting-with-machine-learning-using-python-55e90c346647

*  В остальных случаях, ввиду незначительного количества пропусков после проведённой указанной выше операции, было принято решение их исключения из итогового датафрейма

*  !!! Было принято решение об увеличении количества рассматриваемых лет ввиду нехватки данных в изначально заявленных данных (общее число показателей составляло порядка 700 наблюдений, что противоречит изначальному условию проекта).
